```python
# Load Cleaned Data
import pandas as pd

df=pd.read_csv('cleaned_data.csv')
```

```python
# add a new column 'address_length' that contains the length of each address
df['address_length'] = df['address'].apply(lambda x: len(str(x)))
```

```python
# print the sample data after adding the 'address_length' column
print("Sample data after adding 'address_length' column:")
print(df[['address', 'address_length']].head(n=15))
```

```python
# define the bins and labels
bins = [0, 50000, 70000, 100000]
labels = ['low', 'medium', 'high']
```

```python
# Categorize 'salary' into 'salary_category' based on the defined bins and labels
df['salary_category'] = pd.cut(df['salary'], bins=bins, labels=labels, include_lowest=True)
```

```python
# Print the sample data after adding the 'salary_category' column
print("Sample data after adding 'salary_category' column:")
print(df[['salary', 'salary_category']].head(n=15))
```

```python
# Group by 'department' and calculate average salary and age
summary_report = df.groupby('department').agg({
    'salary': 'mean',
    'age': 'mean'
}).reset_index()
```

```python
# Rename columns for clarity
summary_report.columns = ['department', 'average_salary', 'average_age']
```

```python
# Print the summary report
print("Summary Report of average salary and age by department:")
print(summary_report)

# Save the final transformed data and summary report to CSV files
df.to_csv('transformed_data.csv', index=False)
```
