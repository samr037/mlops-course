```python
# Load data from a CSV file into a DataFrame
df = pd.read_csv('mock_data.csv')
# df = pd.read_csv(f'{path}/vgsales.csv')
```

```python
# check the ununique values in a specific column
df['age'].unique()
```

```python
# Step 1: handle missing values
print("Filling missing values in each column with the median of that column.")
df['age'].fillna(df['age'].median(), inplace=True)
df['salary'].fillna(df['salary'].median(), inplace=True)


# Filling missing values in department with unknown
df['department'].fillna('Unknown', inplace=True) 

# Print the sample data after handling missing values
print("\nSample data after handling missing values:")
print(df.head())
```

```python
# Convert 'profile' from JSON string to dictionary (handling empty strings)
import json
df['profile'] = df['profile'].apply(lambda x: json.loads(x) if pd.notnull(x) else {})
```

```python
print("\nSample data after converting 'profile' column:")
print(df[['profile']].head())
```
